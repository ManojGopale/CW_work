```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np

xShape = project.traces[0].wave.shape[0]
x=np.linspace(0,xShape-1, xShape)

fig, axs = plt.subplots(4, sharex=True, figsize=(30,10))
for idx in range(4):
    axs[idx].plot(x,project.traces[idx].wave)
    axs[idx].set_title(project.traces[idx].key)

```

```python
ct = [0x22, 0x4b, 0x5e, 0x09, 0x15, 0x2b, 0x41, 0x5d, 0x12, 0x41, 0x0c, 0x53, 0x0c]
key = [0x61, 0x2a, 0x35, 0x6c, 0x35, 0x5f, 0x29, 0x38, 0x32, 0x2d, 0x6d, 0x24, 0x62]

decoded = ""

for i in range(0, len(ct)):
    pt = ct[i] ^ key[i]
    decoded += chr(pt)

print(f"{decoded}")
```
